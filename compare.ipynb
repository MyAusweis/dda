```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from torchvision import transforms
from torchvision.datasets import CIFAR10

from dda import functional, pil

def show_tensor(data: torch.Tensor):
    plt.imshow(np.transpose(data.numpy(), (1, 2, 0)), interpolation='nearest')

def show_pil(data):
    plt.imshow(np.asarray(data), interpolation='nearest')

def show(tensor, img):
    tensor = np.transpose(tensor.squeeze().numpy(), (1,2,0))
    img = np.asarray(img) / 255
    plt.imshow(np.concatenate([tensor, img], axis=1))

dataset = CIFAR10("~/.torch/data/cifar10",
                 transform=transforms.Compose([
                    transforms.ToTensor(),
                ]))

raw_dataset = CIFAR10("~/.torch/data/cifar10",
                 transform=transforms.Compose([
                ]))

data = dataset[0][0]
raw_data = raw_dataset[0][0]
```

```python
show(data, raw_data)
```

```python
show(functional.shear_x(data, torch.tensor([0.3])), pil.shear_x(raw_data, -0.3, False))
```

```python
show(functional.translate_x(data, torch.tensor([0.3])), pil.translate_x(raw_data, -0.3, False))
```
