```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from torchvision import transforms
from torchvision.datasets import CIFAR10
from torchvision.utils import make_grid

def show(data: torch.Tensor,
         denormalize=False):
    if denormalize:
        mean = torch.ones(1) * 0.5
        std = torch.ones(1) * 0.5
        data.mul_(std[:, None, None]).add_(mean[:, None, None])
    plt.imshow(np.transpose(make_grid(data).data.numpy(), (1, 2, 0)), interpolation='nearest')

dataset = CIFAR10("~/.torch/data/cifar10",
                 transform=transforms.Compose([
                    transforms.ToTensor(),
                ]))

batch = torch.stack([d for d, _ in [dataset[i] for i in range(10)]])

show(batch)
```
